```python
#Crear una tarjeta personal. El usuario (mediante input) puede elegir:
# Crear datos nuevos
# Modificar datos
# Eliminar datos
datos = {}
while True:
    print("----------\nMenu\n1.Crear datos\n2.Modificar datos\n3.Eliminar Datos\n4.Salir\n5.Datos")
    numero = input("INGRESE UN NUMERO(1-4): ")
    if numero == "1" :
        clave = input("INGRESE CUAL SERA EL DATO(CLAVE): ")
        datos[clave] = input("INGRESE EL VALOR DEL DATO: ")
        print(datos)
    elif numero =="2":
        print("Valores que podes moficar:\n", datos.items())
        mod = input("Cual valor quieres modificar: ")
        valor = input(f"MODIFICA EL {mod}: ")
        datos[mod] = valor
        print(datos)
    elif numero == "3":
        print("Claves que podes eliminar:\n", datos.keys())
        elim = input("INGRESE LA CLAVE PARA ELIMINAR: ")

        for x in datos.keys():
            if [elim] == x:
                del datos[elim]
            else:
                print("LA CLAVE NO EXISTE")
                break
    elif numero == "5":
        print(datos)
    else:
        print("chauuu")
        break
```

```python
class Coche:
    tipo = "Vehiculo"

    #Atributos de instancia
    def __init__(self, modelo, posicion=0):
        self.modelo = modelo
        self.posicion = posicion
    #Definir otras accciones de la clase

    def acelerar(self, metros):
        self.posicion += metros
        print(f"El vehiculo modelo {self.modelo} acelera y avanza {metros} mts. Nueva posicion {self.posicion}")
    
    def girar(self, direccion):
        print(f"El {self.modelo} avanza y gira a la {direccion}")

coche1 = Coche("Toyota", 5)
coche1.acelerar(200)
coche1.girar("izquierda")
```

```python
# Crear dos métodos para nuestro animal y ejecutar, debe ser algo específico de ese animal.
# Ejemplo: "Cuando hablo, digo X".
# Ejemplo: "Cuando me muevo, hago X".
class animal:
    especie = "perro"
    def __init__(self, nombre, color):
        self.nombre = nombre 
        self.color = color

    def sonido(self, sonido):
        self.sonido = sonido
        print(f"Cuando hablo, digo {sonido}")

    def patas(self, patas):
        self.patas = patas 
        print(f"Camino en {patas} patas ")
rompe = animal("Rompe", "Marron")
print(f"Nombre: {rompe.nombre}\nColor: {rompe.color}")
rompe.sonido("guau guau")
rompe.patas(4)
```

```python
class AnimalDeGranja:
    #atributos de clase
    tipo = "animal de granja"
    #atributos de instancia
    def __init__(self, nombre, color, sonido):
        self.nombre = nombre
        self.color = color
        self.sonido = sonido
        #Metodos(Acciones de la clase)
    def hablar(self):
        print(f"Hablo y digo {self.sonido}")

    def presentarse(self):
        print(f"Hola soy {self.nombre}")

class Gallina(AnimalDeGranja):
    huevos_por_dia = 1
    def poner_huevo(self):
        print(f"Estoy poniendo {self.huevos_por_dia}")

class Vaca(Gallina):
    patas = 4
    def mujir(self):
        print("muuuuuu")

gallina1 = Gallina("Carlos", "marron", "cocorocoo")
gallina1.hablar()
gallina1.presentarse()

vaca = Vaca("lola", "negro", "muuuu")
vaca.presentarse()
vaca.hablar()
vaca.mujir()
vaca.poner_huevo()    
```

```python
# Definir una clase madre que herede a sus clases hijas dos atributos de instancia.
# Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.
# Observación: Editar las clases hijas para que no se inicialicen.
class Madre:
    def __init__(self, nombre, color_de_ojo):
        self.nombre = nombre
        self.color_de_ojo = color_de_ojo
        
class Hija(Madre):
    def nombree(self):
        print(f"Mi nombre es {self.nombre} ")
    
    def color_ojo(self):
        print(f"El color de mi ojo es {self.color_de_ojo}")

hija1 = Hija("Ana", "marron")
hija2 = Hija("Belen", "azul")
hija1.nombree()
hija1.color_ojo()
```
