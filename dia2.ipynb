```python
'''
Escribe un programa que pregunte al usuario por su edad y determine 
si tiene edad suficiente para comprar y consumir alcohol.
Si el usuario tiene mas de 18 pero menos de 21 impirmir "Podes tomar pero no comprar"
Si el usuario tiene menos de 18 imprimir "Juguito nomas"
Si el usuario tiene mas de 21 imprimir "Hace lo que quieras"
'''
edad = int(input("Cuantos años tienes"))
if 18< edad <21:
    print("Podes tomar pero no comprar")
elif edad<=18:
    print("Juguito nomas")
elif edad>=21:
    print("Hace lo que quieras")
```

```python
'''
Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), 
imprima la temperatura escrita, y luego clasificar la temperatura ingresada en los siguientes niveles:
- Entre -10 y 18 grados, que imprima "Fresquete".
- Entre 19 y 30 grados, que imprima "Calor'i".
- Entre 31 y 45 grados, que imprima "Hakuuuuuu".
- Para todo lo demás, imprimir "Ñamano mba'e!".
'''
temperatura = float(input("Ingrese la temperatura en grados celsius"))
print(f"La temperatura es de: {temperatura}°C")
#INTERVALO ENTRE -10 Y 18
if -10<=temperatura<=18:
    print("fresquete")
#INTERVALO ENTRE 19 Y 30
elif 19<=temperatura<=30:
    print("Calor'i")
#INTERVALO ENTRE 31 Y 45
elif 31<=temperatura<=45:
    print("hakuuuu")
else:
    print("Ñamano mba'e!")
```

```python
#Escribir un programa que almacene una contraseña en una variable,
#y pregunte al usuario por la contraseña hasta que introduzca la contraseña correcta.
#Cuando el usuario ingrese la contrasena correcta, imprimir un mensaje secreto.

contraseña = "contrase"
while True:
    contraseña_usuario = input("INTRODUSCA LA CONTRASEÑA")
    if contraseña_usuario == contraseña:
        print("Este es el mensaje oculto")
        break
    else:
        print("Contraseña incorrecta")
```
