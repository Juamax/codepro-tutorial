```python

def caracteristicas_de_un_celular(ram, procesador, bateria):
    print(f"La RAM tiene un espacio de: {ram}")
    print(f"El procesador del celular es: {procesador}")
    print(f"La bateria tiene una capacidad de: {bateria}")

ram = input("Cuanto de ram tiene el celular:")
procesador = input("Cual es el procesador del celular:")
bateria = input("De cuanto es la duracion de la bateria:") 

caracteristicas_de_un_celular(ram, procesador, bateria)
```

```python
producto = input("Ingrese el producto a comprar")

def caracteristicas_del_producto():
    primera_caracteristca = input("Ingrese la primera caracteristica del producto: ")
    segunda_caracteristca = input("Ingrese la segunda caracteristica del producto: ")
    tercera_caracteristca = input("Ingrese la tercera caracteristica del producto: ")
    print(F"El producto es: {producto}\nTres caracteristicas del producto son:")
    print(f"{primera_caracteristca}\n{segunda_caracteristca}\n{tercera_caracteristca}")
caracteristicas_del_producto()
```
